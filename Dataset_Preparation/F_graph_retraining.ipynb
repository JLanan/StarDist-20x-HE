```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
path_to_data = r"\\10.99.68.53\Digital pathology image lib\_Image libraries for training\2023-05-09 Published HE Nuclei Datasets\Segmentation Models\StarDist\Retraining Tissue Specific for Graph.csv"
df = pd.read_csv(path_to_data)
df
```

```python
plt.figure(figsize=(10, 6))
for col in df.columns[1:]:
    plt.plot(df[df.columns[0]], df[col], label=col)
plt.xlabel(df.columns[0])
plt.ylabel('PQ')
plt.title('Plot of Series Values')
plt.legend()
plt.grid(True)
plt.show()
```

```python
series_groups = [df.columns[i:i+3] for i in range(1, 10, 3)]

# Plot each group of series
for i, series_group in enumerate(series_groups):
    plt.figure(figsize=(10, 6))  # Adjust the figure size as needed

    # Plot each series in the group
    for col in series_group:
        plt.plot(df['Epochs'], df[col], label=col)

    plt.xlabel('Epochs')  # Set x-axis label
    plt.ylabel('Values')  # Set y-axis label
    plt.ylim(0.5, 0.7)
    plt.title(f'Touch-up Training for JHUN20 {col.split("_")[0]}')  # Set plot title
    plt.legend()  # Show legend
    plt.grid(True)  # Show grid
    plt.show()
```

```python
# Create a figure and subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6))  # 1 row, 3 columns

# Plot each group of series
for i, series_group in enumerate(series_groups):
    ax = axes[i]

    # Plot each series in the group
    for col in series_group:
        ax.plot(df['Epochs'], df[col], label=col)
    
    ax.legend()
    ax.grid(True)
    ax.set_xlabel('Epochs')
    ax.set_ylim(0.5, 0.7)
    
    if i == 0:
        ax.set_ylabel('PQ  (=0.5)')
    elif i == 1:
        ax.set_title('JHUN20 Touch-up Training from SD_HE_20x')
        ax.tick_params(axis='y', labelleft=False)
    else:
        ax.tick_params(axis='y', labelleft=False)

plt.tight_layout()
plt.savefig('Triple Chart.png', bbox_inches='tight')
plt.show()
```
