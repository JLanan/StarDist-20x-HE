```python
from csbdeep.utils import normalize
from skimage.io import imread, imsave, imshow
import numpy as np
import os
import matplotlib.pyplot as plt

img_folder = r"\\10.99.68.53\Digital pathology image lib\_Image libraries for training\2023-05-09 Published HE Nuclei Datasets\20x Native\JHU\images"
img_name = 'FallopianTube_test_3.tif'
img = imread(os.path.join(img_folder, img_name))
imshow(img)
plt.axis('off')
```

```python
img = normalize(img, 3, 99)
img = (np.clip(img, 0, 1) * 255).astype(np.uint8)
imshow(img)
plt.axis('off')
imsave(os.path.join(img_folder, 'norm1.tif'), img)
```
